## Step 4: Visualize the data

Use the next code cell to create a figure that tells a story behind your dataset.  You can use any chart type (_line chart, bar chart, heatmap, etc_) of your choosing!

````python
# Create a plot
# A regression plot to confirm correlation between population and fertility rate

ax = sns.regplot(x = 'Fert. Rate', y= 'Population (2020)', data =df1)

# Check that a figure appears below
step_4.check()
````

````python
sns.pairplot(df1, hue= 'Fert. Rate')
````

````python
ax = sns.regplot(y= 'Population (2020)', x = 'Migrants (net)', data =df1)
````

````python
#This gives the output of the countries with the highest population in the world
highest_pop = my_data.nlargest(n=10, columns=['Population (2020)'])
highest_pop
````

````python
#This gives the output of the countries with the lowest population in the world
least_pop = my_data.nsmallest (n=10, columns=['Population (2020)'])
least_pop
````

````python
df2 = least_pop.sort_values('Population (2020)', ascending = False)
df2
````

````python
y = highest_pop['Country/Other']
x = highest_pop['Population (2020)']
plt.figure(figsize=(10, 6))
ax = sns.barplot(x ,y, data= highest_pop)
plt.title('Ten most populous countries [2020]')
plt.xlabel('Population')
plt.ylabel('Countries')
ax.bar_label(ax.containers[0])

plt.show();
````

````python
y = df2['Country/Other']
x = df2['Population (2020)']
plt.figure(figsize=(10, 6))
ax = sns.barplot(x ,y, data= df2)
plt.title('Ten least populous countries [2020]')
plt.xlabel('Population')
plt.ylabel('Countries')
ax.bar_label(ax.containers[0])

plt.show()
````

### Overview:

- Using the population figures of 2020. The following can be extrapolated from the graph above:
- The two most populous countries on the earth are China and India with a population of 1.439 and 1.38 billion respectively.
- Other most populous countries with millions of residents include United States, Indonesia, and Pakistan with 3.31, 2.73, and 2.2 million people respectively.
- Two countries, Antigua and Barbuda and Seychelles are the least populated countries in the world with populations less than a hundred thousand.
   

````python
max_den = my_data['Density (P/Km²)'].max()
print(max_den)
min_den = my_data['Density (P/Km²)'].min()
print(min_den)
````

````python
highest_den = my_data.nlargest(n=10, columns=['Density (P/Km²)'])
highest_den
````

````python
least_den = my_data.nsmallest (n=10, columns=['Density (P/Km²)'])
least_den
````

````python
least_den2 = least_den.sort_values('Density (P/Km²)', ascending = False)
least_den2
````

````python
plt.figure(figsize=(10, 6))
x = highest_den['Country/Other']
y = highest_den['Density (P/Km²)']
ax = sns.barplot(x ,y, data= highest_den)
plt.title('Most denselypopulated countries [2020]')
plt.ylabel('Density (P/Km²)')
plt.xlabel('Countries')
plt.xticks (rotation = -45)
ax.bar_label(ax.containers[0])

plt.figure(figsize=(10, 6))
x = least_den2['Country/Other']
y = least_den2['Density (P/Km²)']
ax = sns.barplot(x ,y, data= least_den2)
plt.title('Least densely populated countries [2020]')
plt.ylabel('Density (P/Km²)')
plt.xlabel('Countries')
plt.xticks (rotation = -45)
ax.bar_label(ax.containers[0])

plt.show();
````

### Overview:

- Macau is the most densely populated country with 21645 people per square kilometer which is almost three times the next most densely population, Monaco (8356).
- Western Sahara and Mongolia are the least densely populated countries with 2 people per square kilometer followed by Namibia, Iceland, and Australia with 3 people persquare kilometer.
